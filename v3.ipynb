```python
import xlrd

# Read data file
data = xlrd.open_workbook("simplified_subway_info.xlsx")
data = data.sheet_by_name('Sheet1')

# Store the loaded book object as a string list in subwayStation variable
subwayStation = []
for line in range(4) :
    cur = [x for x in data.col_values(line) if x]    
    subwayStation.append(cur[1:])
```

```python
class SubwayLine :

    def __init__ (self, stations) :
        self.stations = stations
    
    def __str__ (self) :
        return str((self.stations))

    def __and__ (self, other) :
        commonStation = tuple(set(self.stations) & set(other.stations))

        if len(commonStation) > 0:
            return commonStation
            
        else :
            return None

    def __contains__ (self, station) :
        if station in self.stations :
            return True

        else :
            return False

    def getpath (self, dpt, dst) :
      if dpt and dst in self.stations :
        dptIdx = self.stations.index(dpt)
        dstIdx = self.stations.index(dst)
        path = list(self.stations[min(dptIdx, dstIdx) : max(dptIdx, dstIdx) + 1])
        if dptIdx > dstIdx :
          path.reverse()
        return path
      else :
        return None


SubwayLine_1 = SubwayLine(subwayStation[0])
SubwayLine_2 = SubwayLine(subwayStation[1])
SubwayLine_3 = SubwayLine(subwayStation[2])
SubwayLine_4 = SubwayLine(subwayStation[3])

'''
print(SubwayLine_1.stations)
print(SubwayLine_1 & SubwayLine_2)
print('동대문' in  SubwayLine_1.stations)
print(SubwayLine_1.getpath('서울역','신도림'))
'''
```
